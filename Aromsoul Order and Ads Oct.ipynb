```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel(r"DOWNLOADS/Aromsoul RP Oct Month.xlsx")
df.head(1)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
ms=df.groupby("product-name")["quantity"].sum().sort_values(ascending=False).reset_index()

ms=ms.head(400)
ms
```

```python
# Create a Plot
#plt.figure(figsize=(8,6))
a=sns.barplot(data=ms,x="quantity",y="product-name",ci=None)
ax1=plt.subplot(111)

# Sort Values 
ms.sort_values("quantity",ascending=False)
plt.xlabel('Quantity', fontsize=18)
plt.ylabel('Product Name', fontsize=16)
plt.title("Top Selling Products of Aromsoul",fontsize=20)

#Annotate Values
ax1.xaxis.grid(linestyle="--",linewidth=0.5)
for pY , pX in enumerate(ms["quantity"].head(10)):
    ax1.annotate(pX, xy=(pX, pY),fontstyle="italic", va="center")
ax1.set_xlim(0,ms["quantity"].max()*1.1)
plt.show()
```

```python
mon=df[df["day"]=="Sunday"]
mon=mon[["product-name","quantity"]]
mon=mon.groupby("product-name")["quantity"].sum().sort_values(ascending=False).to_frame().reset_index()
mon
```

```python
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
july=df.groupby("day")["quantity"].sum().reindex(cats).reset_index()
july.columns=("Day","Quantity")



fig, ax1 = plt.subplots(figsize=(10,6))
color = 'tab:green'
#bar plot creation
ax1.set_title('October Analysis', fontsize=16)
ax1.set_xlabel('Month', fontsize=16)
ax1.set_ylabel('Avg Temp', fontsize=16)
ax1 = sns.barplot(x='Day', y='Quantity', data =july, palette='summer')
ax1.tick_params(axis='y')
#specify we want to share the same x-axis
ax2 = ax1.twinx()
color = 'tab:red'
#line plot creation
ax2.set_ylabel('Avg Percipitation %', fontsize=16)
ax2 = sns.lineplot(x='Day', y='Quantity', data = july, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
#show plot
plt.show()
```

```python
df['time'] = df['time'].apply(str)
```

```python
mon=df[df["day"]=="Sunday"]
mon=mon[["product-name","quantity","time"]]
#mon=mon.groupby("product-name")[["time","quantity"].sum().sort_values(by="time",ascending=False).to_frame().reset_index()
mon.groupby("time")[["product-name","quantity"]].sum().sort_values(by="quantity",ascending=False).reset_index()
```

```python
df["Revenue"]=df["quantity"]*(df["item-price"])
revenue=df.groupby('product-name')[["quantity","Revenue"]].sum().sort_values(by="Revenue",ascending=False).astype(int).reset_index()
revenue
```

```python
most_saa=df.groupby("ship-state").sum().sort_values("quantity", ascending=False).reset_index()
most_saa=most_saa[["ship-state","quantity"]]


most_saa=most_saa.head(6)
most_saa
```

```python
#plt.figure(figsize=(10,8))
june=sns.barplot(data=most_saa,x="ship-state",y="quantity")
plt.xlabel('State', fontsize=18)
plt.ylabel('Quantity', fontsize=16)
plt.title("Top Selling States",fontsize=20)
june.set_xticklabels(june.get_xticklabels(),rotation=45,horizontalalignment="right")
plt.show()

```

```python

maharashtra=df[df["ship-state"]=="KARNATAKA"]
maharashtra=maharashtra.groupby('product-name')["quantity"].sum().sort_values(ascending=False).reset_index()
maharashtra=maharashtra.head()
maharashtra
```

```python
# Create a Plot
#plt.figure(figsize=(8,6))
a=sns.barplot(data=maharashtra,x="quantity",y="product-name",ci=None)
ax1=plt.subplot(111)

# Sort Values 
maharashtra.sort_values("quantity",ascending=False)
plt.xlabel('Quantity', fontsize=18)
plt.ylabel('Product Name', fontsize=16)
plt.title("Top Selling Products in Karnataka",fontsize=20)

#Annotate Values
ax1.xaxis.grid(linestyle="--",linewidth=0.5)
for pY , pX in enumerate(maharashtra["quantity"]):
    ax1.annotate(pX, xy=(pX, pY),fontstyle="italic", va="center")
ax1.set_xlim(0,maharashtra["quantity"].max()*1.1)
plt.show()
```
